```python
import json
import os

# folder = os.listdir('')
file_path = './data/train_data.jsonl'

with open(file_path, 'r') as file:
    data = [json.loads(line) for line in file]
```

```python
import random
from tqdm import tqdm

with open('./data/law.json', 'r') as file:
    l_data = json.load(file)

def search_content(law):
    num = law.split("第")[1].split("條")[0]
    l = law.split("第")[0]
    
    if l in l_data.keys():
        law_dict = l_data[l]
        list_law_dict= list(law_dict.keys())
        ids = (list_law_dict).index(f'第 {num} 條')
        pre_range = max(0, ids-5)
        post_range = min(len(list_law_dict), ids+5)
        # print(ids, pre_range, post_range)
        content = []
        for c_i in range(pre_range, post_range):
            name = list_law_dict[c_i].replace(' ','')
            content.append(f'{l}{name}:'+l_data[l][list_law_dict[c_i]].replace(' ',''))       
        random.shuffle(content)
        # print(content)
        
        n_content = []
        
        cont = 0
        while(1):
            cont+=1
            random_idx = random.randint(0, len(list_law_dict))
            random_pre_range = max(0, random_idx-2)
            random_post_range = min(len(list_law_dict), random_idx+2)
            if ids> random_post_range or ids<random_pre_range:
                break
            if cont >10:
                return '/n'.join(content)
            
        for c_i in range(random_pre_range, random_post_range):
            name = list_law_dict[c_i].replace(' ','')
            n_content.append(f'{l}{name}:'+l_data[l][list_law_dict[c_i]].replace(' ','')) 

        return '/n'.join(content), '/n'.join(n_content)
        
    else:
        return 'N/A'

dict = []

for idx, d in enumerate(tqdm(data)):
    label = d['label'].split(',')
    try:
        scene = d['title']+ d['question']
    except:
        print(d['title'], d['question'])
        scene = d['title']
    for l in label:
        try:
            content = search_content(l)
        except:
            continue 
        if content == 'N/A':
            continue
        # else:
        #     print(len(content))
        # print(l)
        # print(content)
        tmp_dict = {}  
        tmp_dict['instruction'] = f'參考台灣部份法律條文，判斷以下情境，是否有相關條文，與那條法律條文最相關'
        tmp_dict['input'] = f'法律條文:{content[0]}, \n情境：{scene}'
        tmp_dict['output'] = '是,'+l
        # dict.append(tmp_dict)
        neg_tmp_dict = {}  
        neg_tmp_dict['instruction'] = f'參考台灣部份法律條文，判斷以下情境，是否有相關條文，與那條法律條文最相關'
        neg_tmp_dict['input'] = f'法律條文:{content[1]},\n情境： {scene}'
        neg_tmp_dict['output'] = '否,無相關'
        dict.extend([tmp_dict, neg_tmp_dict])
    # print(tmp_dict)
    print(d['id'])
    # break
```
